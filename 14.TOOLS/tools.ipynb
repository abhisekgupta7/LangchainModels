```python
from langchain_core.tools import Tool
```

```python
def multiply(a,b):
    """Multiplies two numbers."""
    return a * b
```

```python
def multiply(a: int, b: int) -> int:
    """Multiplies two numbers."""
    return a * b
```

```python
@tool 
def multiply(a: int, b: int) -> int:
    """Multiplies two numbers."""
    return a * b
```

```python
result = multiply.invoke({"a": 3, "b": 4})
```
